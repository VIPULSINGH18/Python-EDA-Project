```python
#in this code we have converted string data type to date time datatype
df['timeStamp']=pd.to_datetime(df['timeStamp'])
type(df['timeStamp'].iloc[0])
```

```python
# we can access different attributes of date time object

time= df['timeStamp'].iloc[0]
time

# time.hour
# time.year
# time.month
#time.week
#time.day
#time.dayofweek # monday coded as 0 and then sunday as 6..
```

```python
df['hour']= df['timeStamp'].apply(lambda time:time.hour)
df['hour']
df['month']= df['timeStamp'].apply(lambda time:time.month)
df['month']
df['Day of week']= df['timeStamp'].apply(lambda time:time.dayofweek)
df['Day of week']
df.head()
```

```python
# to assign name of day instead of numbers
dmap = { 0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
df['Day of week']= df['Day of week'].map(dmap)
df.head()

```

```python
#to create countplot of reason of accident weekly wise 
sns.countplot(x='Day of week',data=df,hue='reason', palette='viridis')
plt.show()
#to relocate the position of legend
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
# creating 3 different plots of 3 reasons ( ems,fire,traffic) where we plot the count of twp per month for different reasons


df[df['reason']=='Traffic'].groupby('date').count()['twp'].plot()   # we used groupby('date').count() in place of bydate, because in starting we used df and we dont have any attribute named bydate inside df ...
plt.tight_layout()
plt.title('Traffic')
plt.show()
```

```python
df[df['reason']=='Fire'].groupby('date').count()['twp'].plot()  
plt.tight_layout()
plt.title('Fire')
plt.show()
```

```python
df[df['reason']=='EMS'].groupby('date').count()['twp'].plot()  
plt.tight_layout()
plt.title('EMS')
plt.show()
```

```python
# if we pass 2 attributes in group by method without unstack command then it becomes multi indexing.. 
G1= df.groupby(by=['Day of week','hour']).count() # without unstack command 
G1
```

```python
# if we pass 2 attributes in group by method with unstack command then first attribute become row and second becomes column...

G1= df.groupby(by=['Day of week','hour']).count().unstack() #with unstack command
G1
```

```python
G1= df.groupby(by=['Day of week','hour']).count()['reason'].unstack() #with unstack command
G1
```

```python
#creating a heatmap using seaborn of G1 data...
plt.figure(figsize=(12,6))  # this command is used to adjust the size of heatmap...
sns.heatmap(G1,cmap='viridis')
plt.show()
```

```python
# using clustermap for G1 data...

sns.clustermap(G1,cmap='viridis')
plt.show()
```
